```python
#Pergunta 1

def calcular_nota_final(teste, trabalho):
    if not (0 <= teste <= 20) or not (0 <= trabalho <= 20):
        return "Notas inválidas. As notas devem estar entre 0 e 20"
    
    mensagem = "Aluno precisa fazer o exame final;"
    motivo_encontrado = False
    if teste < 8:
        mensagem += " motivo: nota do teste inferior a 8."
        motivo_encontrado = True
    if trabalho < 8:
        mensagem += " motivo: nota do trabalho inferior a 8."
        motivo_encontrado = True
    if motivo_encontrado:
        return mensagem
    
    nota_final = teste * 0.45 + trabalho * 0.55
    
    if nota_final >= 10:
        return f"Aluno aprovado. Nota final: {nota_final:.2f}"
    else:
        return f"Aluno reprovado. Nota final: {nota_final:.2f}"
```

```python
calcular_nota_final(10,10)
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Nome do ficheiro
ficheiro = 'ActividadeFisica.xlsx'

# 1. Verificar se o ficheiro existe
if not os.path.exists(ficheiro):
    print(f"Erro: O ficheiro '{ficheiro}' não foi encontrado.")
else:
    # 2. Ler o ficheiro Excel
    try:
        df = pd.read_excel(ficheiro)
    except Exception as e:
        print(f"Erro ao ler o ficheiro: {e}")
    else:
        # 3. Verificar se o ficheiro contém dados
        if df.empty:
            print("O ficheiro está vazio.")
        else:
            # 4. Converter a coluna 'Passos' para um array NumPy
            passos_array = df['Passos'].to_numpy()

            # 5. Calcular o total de passos
            total_passos = np.sum(passos_array)
            print(f"Total de passos dados: {total_passos}")

            # 6. Representação gráfica
            plt.figure(figsize=(10, 6))
            plt.bar(df['Data'].astype(str), df['Passos'], color='skyblue')
            plt.title('Nº de Passos por Dia')
            plt.xlabel('Data')
            plt.ylabel('Passos')
            plt.xticks(rotation=45)
            plt.grid(axis='y', linestyle='--', alpha=0.7)
            plt.tight_layout()
            plt.show()
```
